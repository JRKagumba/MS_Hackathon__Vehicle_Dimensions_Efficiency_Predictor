# **01 - Data Collection**

## Overview

This document outlines the first phase of the data collection process for our machine learning project aimed at determining a vehicle's energy efficiency based on its dimensions. The objective of this phase was to gather relevant data from reliable sources and ensure it was accurately loaded into our data management system for further analysis.

## Data Sources

Our data was collected from two primary sources known for their comprehensive vehicle data:

1. **FuelEconomy.gov Web Services** - This site provides a broad range of data on vehicle fuel economy. It includes information on makes, models, and their energy efficiency metrics among other details, which are crucial for our analysis. Data was accessed via their web services API.
   - Access URL: [https://www.fueleconomy.gov/feg/ws/](https://www.fueleconomy.gov/feg/ws/)

2. **Canadian Vehicle Specifications** - Offered by the Canadian Association of Road Safety Professionals, this source provides detailed specifications on vehicles available in Canada, including safety features and dimensions. This dataset complements the energy efficiency data by providing additional context on vehicle size and structure.
   - Access URL: [https://carsp.ca/en/news-and-resources/advanced-vehicle-safety-systems/canadian-vehicle-specifications/](https://carsp.ca/en/news-and-resources/advanced-vehicle-safety-systems/canadian-vehicle-specifications/)

## Data Loading Process

After acquiring the data from the aforementioned sources, the files were first downloaded locally. This step was necessary to perform an initial inspection and ensure the integrity of the data before uploading it to our data management system. Following this verification, the data was manually loaded into an MS Fabric Lakehouse, specifically into the following path for raw data storage:

- **Lakehouse Path**: `/lakehouse/default/Files/data/raw`

This centralized repository within our Lakehouse environment is designated for storing raw data collected from external sources. The manual loading process involved careful organization to maintain the structure and accessibility of the data for subsequent processing and analysis phases.

## Next Steps

With the data successfully collected and stored, the next phase involves data cleaning and preprocessing. This will prepare our dataset for the machine learning model development, where we will analyze the relationship between vehicle dimensions and their energy efficiency.
